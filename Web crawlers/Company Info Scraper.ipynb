```python
import requests
from bs4 import BeautifulSoup
import urllib
import time
import os
import re
import pandas as pd
```

```python
base_url = 'https://www.malaysiastock.biz/Market-Watch.aspx?type=A&value={}'
df = pd.DataFrame(columns=['company_full_name', 'company_symbol', 'last_quarter', 
                           'last_quarter_revenue_000', 'last_quarter_net_profit_000',
                           'board', 'shariah_compliant', 'sector', 'sub_sector', 
                           'market_capital', 'num_of_share', 'roe'])

for c in "ABCDEFGHIJKLMNOPQRSTUVWXYZ0":
    url = base_url.format(c)
    headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X '
               '10_14_3) AppleWebKit/537.36 (KHTML, like Gecko) '
               'Chrome/72.0.3626.109 Safari/537.36'}
    html_doc = requests.get(url, headers=headers).text
    soup = BeautifulSoup(html_doc, 'html.parser')
    company_links = soup.select("tr > td.alignLeft > span > a")
    for co in company_links:
        co_url = "https://www.malaysiastock.biz/" + co.get('href')
        co_name = co.text
        print('Scraping', co_name, co_url)
        co_html_doc =  requests.get(co_url, headers=headers).text
        co_soup = BeautifulSoup(co_html_doc, 'html.parser')
        try:
            last_quarter = co_soup.select_one(
                '#MainContent_gvReport tr:nth-of-type(2) > td:nth-of-type(4)').text
            last_quarter_revenue = co_soup.select_one(
                '#MainContent_gvReport tr:nth-of-type(2) > td:nth-of-type(5)').text
            last_quarter_net_profit = co_soup.select_one(
                '#MainContent_gvReport > tr:nth-of-type(2) > td:nth-of-type(7)').text
            symbol = co_soup.select_one('#ctl13_lbSymbolCode').text.split(' (')[0]
            fullname = co_soup.select_one('#ctl13_lbCorporateName').text
            board = co_soup.select_one('#ctl13_lbMarket > span').text
            shariah_compliant = co_soup.select_one('#ctl13_lbShariah > span').text
            sector = co_soup.select_one('#ctl13_lbSector > a').text
            sub_sector = co_soup.select_one('#ctl13_lbSubSector > a').text
            market_capital = co_soup.select_one('#MainContent_lbFinancialInfo_Capital').text
            num_of_share = co_soup.select_one('#MainContent_lbNumberOfShare').text
            roe = co_soup.select_one('#MainContent_lbFinancialInfo_ROE').text
            df.loc[df.shape[0], :] = (fullname, symbol, last_quarter, last_quarter_revenue,
                                      last_quarter_net_profit, board, shariah_compliant, sector,
                                      sub_sector, market_capital, num_of_share, roe)
        except Exception as e:
            print('='*40)
            print('Failed', co_name, co_url)
            print('='*40)
```
